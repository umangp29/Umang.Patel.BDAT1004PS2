# Question 5

Write a recursive method binary() that takes a non-negative integer n and prints the binary representation of integer n.


```python
def binary(n):
    if n<2:
        a = str(n)
    else:
        a = binary(n // 2) + str(n % 2)
    return a.replace("'"," ")
```

Implement a class named HeadingParser that can be used to parse an HTMLdocument, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your
implementation using w3c.html.

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    inHeading = False
    sec=False
    
    # This code handle all the start tags that are h1 and h2 in this w3c.html webpage.
    def handle_starttag(self, tag, attrs):
        if tag == "h1":
            self.inHeading = True
        if tag=="h2":
            self.sec=True
            
    # This code retreive all data from start tags.
    def handle_data(self, data):
        if self.inHeading:
            print(data)
        if(self.sec):
            print(" "+data)
            
    #This code handle end heading tags.
    def handle_endtag(self, tag):
        if tag =="h1":
            self.inHeading = False
        if tag =="h2":
            self.inHeading = False
```

```python
# Open the file "w3c.html" for reading and read its contents into the "content" variable.
infile=open('w3c.html')
content=infile.read()
infile.close()

# Create an instance of the HeadingParser class and feed it the HTML content.
hp=HeadingParser()
hp.feed(content)
```

```python
import urllib.request
import re

def webdir(url, depth, indent=0):
    # Print the URL with the specified indentation
    print(' ' * indent + url)

    # Stop recursion if (depth = zero).
    if depth == 0:
        return

    # Open the URL and read content from it.
    with urllib.request.urlopen(url) as response:
        content = response.read().decode()

    #regular expression is used to find all links in page.
    link_pattern = re.compile('<a\s+(?:[^>]*?\s+)?href="([^"]*)"')
    links = link_pattern.findall(content)

    # Recursively visit each link with increased indentation reduced depth.
    for link in links:
        if link.startswith('http'):
            webdir(link, depth - 1, indent + 4)
        elif link.startswith('/'):
            base_url = '/'.join(url.split('/')[:3])
            webdir(base_url + link, depth - 1, indent + 4)
```
